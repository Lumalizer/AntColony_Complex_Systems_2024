```python
import pandas as pd
from pyrqa.settings import Settings
from pyrqa.neighbourhood import FixedRadius
from pyrqa.time_series import TimeSeries
from pyrqa.computation import RQAComputation

def load_and_preprocess_data(file_name):
    data = pd.read_csv(file_name)
    # Assuming 'x' and 'y' are the coordinates we're interested in
    return data['x'].values, data['y'].values  # Separate the dimensions for individual processing

def perform_RQA(time_series_data, radius=0.5, embedding_dimension=2, time_delay=1):
    time_series = TimeSeries(time_series_data, embedding_dimension=embedding_dimension, time_delay=time_delay)
    settings = Settings(time_series, neighbourhood=FixedRadius(radius))
    result = RQAComputation.create(settings).run()
    return result

# Iterate over your CSV files
for i in range(1, 11):
    file_name = f'dataset/ants/csvs/ants_seq{i}.csv'
    x_data, y_data = load_and_preprocess_data(file_name)

    # Analyze each dimension separately
    print(f'Results for {file_name}:')
    for dimension, data in zip(['x', 'y'], [x_data, y_data]):
        rqa_result = perform_RQA(data)
        print(f'Dimension {dimension}:')
        print(rqa_result)
```

```python
# analyze on an ant level whether an ants position recurrs with iteself (potentially explaining behavior of a single ant)
import pandas as pd
from pyrqa.time_series import TimeSeries
from pyrqa.settings import Settings
from pyrqa.neighbourhood import FixedRadius
from pyrqa.computation import RQAComputation
from pyrqa.metric import EuclideanMetric
import matplotlib as plt
from pyrqa.computation import RPComputation
from pyrqa.image_generator import ImageGenerator
import io
from pyrqa.analysis_type import Cross
import random
```

```python
# data for a single ant
file_paths = ['dataset/ants/csvs/ants_seq1.csv', 'dataset/ants/csvs/ants_seq2.csv', 'dataset/ants/csvs/ants_seq3.csv', 'dataset/ants/csvs/ants_seq4.csv', 'dataset/ants/csvs/ants_seq5.csv', 'dataset/ants/csvs/ants_seq6.csv', 'dataset/ants/csvs/ants_seq7.csv', 'dataset/ants/csvs/ants_seq8.csv', 'dataset/ants/csvs/ants_seq9.csv', 'dataset/ants/csvs/ants_seq10.csv']
results = {}

for file in file_paths:
    ants = pd.read_csv(file)
    unique_ants = ants['ant_id'].unique()
    for ant_id in unique_ants:
        ant_data = ants[ants['ant_id'] == ant_id].dropna(subset=['x', 'y'])

    # time series
    time_series = TimeSeries(data=ant_data[['x', 'y']].values.flatten(), embedding_dimension=2, time_delay=2)

    settings = Settings(time_series, neighbourhood=FixedRadius(5), similarity_measure=EuclideanMetric, theiler_corrector=1)


    result = RQAComputation.create(settings).run()
    results[(file, ant_id)] = result

for key, value in results.items():
    print(f"Results for {key}, Ant ID {key[1]}, RQA: {value}")
```

```python
# analyze rqa for a random ant in the indoor and outdoor dataset 

files_indoor = ['dataset/ants/csvs/ants_seq1.csv', 'dataset/ants/csvs/ants_seq2.csv', 'dataset/ants/csvs/ants_seq3.csv', 'dataset/ants/csvs/ants_seq4.csv', 'dataset/ants/csvs/ants_seq5.csv']
files_outdoor= ['dataset/ants/csvs/ants_seq6.csv', 'dataset/ants/csvs/ants_seq7.csv', 'dataset/ants/csvs/ants_seq8.csv', 'dataset/ants/csvs/ants_seq9.csv', 'dataset/ants/csvs/ants_seq10.csv']

outdoor_carpet = ['dataset/ants/csvs/ants_seq6.csv', 'dataset/ants/csvs/ants_seq8.csv', 'dataset/ants/csvs/ants_seq9.csv']
outdoor_little = ['dataset/ants/csvs/ants_seq7.csv', 'dataset/ants/csvs/ants_seq10.csv']
```

```python
# rqa for indoor ants
results_indoor = {}
for file in files_indoor:
    ants = pd.read_csv(file)
    unique_ants = ants['ant_id'].unique()
    # chose random unique ants
    ant_id = random.choice(unique_ants)

    ant_data = ants[ants['ant_id'] == ant_id].dropna(subset=['x', 'y'])

    # time series
    time_series = TimeSeries(data=ant_data[['x', 'y']].values.flatten(), embedding_dimension=2, time_delay=2)

    settings = Settings(time_series, neighbourhood=FixedRadius(5), similarity_measure=EuclideanMetric, theiler_corrector=1)

    result = RQAComputation.create(settings).run()
    results_indoor[(file, ant_id)] = result
    print(f"Results for {file}, Ant ID {ant_id}, RQA: {result}")

    # generate a recurrence plot for the ant
    rp_comp = RPComputation.create(settings)
    result_rp = rp_comp.run()
    ImageGenerator.save_unthresholded_recurrence_plot(result_rp.recurrence_matrix_reverse_normalized, f"indoor_{ant_id}.png")
```

```python
# CRQA
# data for two ants from the first file, cRQA can ony be conducted for the indoor condition 
file_path_1 = 'dataset/ants/csvs/ants_seq9.csv'

ants_1 = pd.read_csv(file_path_1)
ant_id_1 = random.choice(ants_1['ant_id'].unique())
ant_id_2 = random.choice(ants_1['ant_id'].unique())

ant_data_1 = ants_1[ants_1['ant_id'] == ant_id_1].dropna(subset=['x', 'y'])
ant_data_2 = ants_1[ants_1['ant_id'] == ant_id_2].dropna(subset=['x', 'y'])

time_series_1 = TimeSeries(data=ant_data_1[['x', 'y']].values.flatten(), embedding_dimension=2, time_delay=2)
time_series_2 = TimeSeries(data=ant_data_2[['x', 'y']].values.flatten(), embedding_dimension=2, time_delay=2)
time_series_comb = (time_series_1,
                     time_series_2)

settings_crqa = Settings(time_series_comb, neighbourhood=FixedRadius(5), similarity_measure=EuclideanMetric, theiler_corrector=1, analysis_type=Cross)

computation_crqa = RQAComputation.create(settings_crqa)
result_crqa = computation_crqa.run()

print(result_crqa)
```

```python
computation_crqa_plot = RPComputation.create(settings_crqa)
result_crqa_plot = computation_crqa_plot.run()

ImageGenerator.save_recurrence_plot(result_crqa_plot.recurrence_matrix_reverse,
                                                  'unthresholded_cross_recurrence_plot.png')
```
