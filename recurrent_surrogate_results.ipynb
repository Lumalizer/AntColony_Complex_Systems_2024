```python
import pandas as pd
import re

# Function to extract values from a single text snippet
def extract_values(text):
    lines = text.strip().split('\n')
    values = {}
    for line in lines:
        match = re.match(r'(.+?):\s+(.+)', line)
        if match:
            key, value = match.groups()
            values[key.strip()] = value.strip()
    return values
```

```python
def get_rqa_measures(texts):
    # Split the long text into individual text snippets
    snippets = re.split(r'Results for Ant ID \d+, RQA: RQA Result:\n===========', texts)[1:]

    # Extract values from all text snippets
    data = [extract_values(snippet) for snippet in snippets]

    # Convert extracted values to a DataFrame
    df = pd.DataFrame(data)

    # Display the DataFrame
    return df
    # Save the DataFrame to a CSV file
    #df.to_csv('extracted_rqa_results.csv', index=False)
```

```python
# concatenate the pandas dataframe rqa_sc0, rqa_sc7, rqa_sc8 on top of each other
rqa_surrs = pd.concat([rqa_sc0, rqa_sc7, rqa_sc8], ignore_index=True)
```

```python
# cast all columns to numeric
rqa_surrs = rqa_surrs.apply(pd.to_numeric, errors='coerce')
rqa_oc0 = rqa_oc0.apply(pd.to_numeric, errors='coerce')
```

```python
# for all columns calculate the mean of the column and the standard deviation of the column but only count the non 'nan' values
import numpy as np
numeric_rqa_oc0 = rqa_oc0.select_dtypes(include=[np.number])
mean_oc0 = numeric_rqa_oc0.mean()
std_oc0 = numeric_rqa_oc0.std()
numeric_rqa_surrs = rqa_surrs.select_dtypes(include=[np.number])
mean_surrs = numeric_rqa_surrs.mean()
std_surrs = numeric_rqa_surrs.std()
```

```python
# do statistical signifance tests to compare the two dataframes for each column calculate the p-value
from scipy.stats import ttest_ind
p_values = {}
for column in numeric_rqa_surrs.columns:
    t, p = ttest_ind(numeric_rqa_surrs[column].dropna(), numeric_rqa_oc0[column].dropna())
    p_values[column] = p
p_values

# calculate T test for each column
t_values = {}
for column in numeric_rqa_surrs.columns:
    t, p = ttest_ind(numeric_rqa_surrs[column].dropna(), numeric_rqa_oc0[column].dropna())
    t_values[column] = t
t_values
```

```python
# create a table with the means and standard deviations of the two dataframes and the p-values
summary = pd.DataFrame({
    'mean_oc0': mean_oc0,
    'std_oc0': std_oc0,
    'mean_surrs': mean_surrs,
    'std_surrs': std_surrs,
    'p_value': p_values,
    't_value': t_values
})
```
